```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
df = pd.read_csv('house_price_area_only.csv')
df.head()
```

```python
df['intercept'] = 1

lm = sm.OLS(df['price'], df[['intercept', 'area']])
results = lm.fit()
results.summary()

# y = mx + b # y = b1x + b0
# m = slope = area = 348.4664
# b = intercept = 9587.8878 i.e the value of y when x = 0

# y = 9588 + 348x i.e for every 1 unit increase(thats x + 1) the value of y = 9588 + 348(+ 348) 
# pvalue is read from the area more

# R-squared explains the percentage variability in price explained by area

# n.b R is the correlation co-efficient while the coefficient of determination is R-squared which explains the percentage of variability
```

```python
df2 = pd.read_csv('house_prices.csv')
df2.head()
```

```python
df2['intercept'] = 1

# n.b dependent variable comes first in the sm.OLS & lm = linear model
lm = sm.OLS(df2['price'], df2[['intercept', 'bathrooms', 'bedrooms', 'area']])
results = lm.fit()
results.summary()
```

```python
X = df2[['intercept', 'bathrooms', 'bedrooms', 'area']]
y = df2['price']


# we can find the optimal b estimates by calculating (XiX)-1Xiy
# where Xi is transposed X and we use np.dot to multiply and linalg.inv to invert to -1
np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
```
